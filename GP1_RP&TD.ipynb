```python
class budget_manager():  
    
    def __init__(self): 
        self.salary=0
        self.expenses=[] 
        self.expense_categories=[]
        self.total_expenses=0
        self.monthly_salary=0
        
        
    def set_salary(self):  # choice 1  
        self.salary=float(input("Enter your yearly salary: "))   # user sets their yeary salary
        self.monthly_salary=self.salary/12  # computes their monthly salary
    
    def set_expenses(self):  # choice 2 
        while True:
            print("\nWelcome to the Expense Tracker")
            print("1: To add a category for expenses and money spend on category")
            print("2: Quit Expense Tracker")
            echoice=int(input("Select a Number (1-2): "))
            # allows user to enter their monthly expenses and split that into categories
            if echoice==1:
                category=input("Enter a category for expenses: ")
                amount=float(input("Enter the amount spent on that category monthly: "))
                self.expenses.append({category:amount}) #uses dictionary inside a list
            elif echoice==2:
                print(self.expenses)  # when user quits expense tracker it prints the list which includes the categories and the amount spent
                break
            else:
                print("Invalid number, please choose a number between (1-2)") # if user enters number other than 1 or 2 it informs them      
    
    def show_expenses(self):  # choice 3 
        for i in self.expenses:
            for v in i.values():
                self.total_expenses+=int(v)  # this computes what their total expenses are
        for i in self.expenses:
            for k in i.keys():
                self.expense_categories.append(k) # adds what the categories are from the expenses
                
        if self.salary!=0 and self.total_expenses!=0:
            print(f"You make ${self.monthly_salary:.2f} per month and your total monthly expenses are ${self.total_expenses}.") # shows monthly salary and that their total expenses are
        print(f"\nYou spent ${self.total_expenses} on the following categories {self.expense_categories}.") # shows total expenses and what categories they spent on
                   
    def sample_budget(self):  # choice 4 
        for i in self.expenses:
                for v in i.values():
                    self.total_expenses+=int(v) 
        if self.salary==0: # if user doesnt enter salary it lets them know so they can quit the sample budget creator
            print("You have forget to enter your salary. Please go back and enter all the information properly so we create a sample budget.") 
        
        print("\nHello, welcome to the Sample Budget Creator\n")
        debt=input("Do you have any debt? (yes/no): ") # asks user if they have debt because it will affect the sample budget
        while True: 
            print("1: Continue with Sample Budget")   # creates a menu for the user to interactive with
            print("2: Quit Sample Budget")
            bchoice=int(input("Select a Number (1-2): "))  # allows user to pick between 1 or 2 to either continue with the creation of their sample budget or let the user quit the program

            if bchoice==1 and debt=='no':  #checks if user has debt and if they want to continue with sample budget
                rent_mortgage_budget = self.monthly_salary * 0.3 # makes salary based off of monthly salary split into 6 generic categories
                utilities_budget = self.monthly_salary * 0.15
                groceries_budget = self.monthly_salary * 0.2
                transportation_budget = self.monthly_salary * 0.15
                entertainment_budget = self.monthly_salary * 0.1
                savings_budget = self.monthly_salary * 0.1

                print("\nSuggested Budget Allocations based off of monthly salary:")  # prints the suggested budget allocations
                print(f"Rent or Mortgage: ${rent_mortgage_budget:.2f}")
                print(f"Utilities (Electricity, Water, Gas): ${utilities_budget:.2f}")
                print(f"Groceries: ${groceries_budget:.2f}")
                print(f"Transportation (Car payments, Fuel, Public Transit): ${transportation_budget:.2f}")
                print(f"Entertainment: ${entertainment_budget:.2f}")
                print(f"Savings: ${savings_budget:.2f}")
                break  # quits the budget creator after it prints the suggestions
                
            elif bchoice==1 and debt=='yes':  #if user has debt it includes that in the budget
                rent_mortgage_budget = self.monthly_salary * 0.25  # percentage of monthly salary that goes towards the budget is decreased because of the debt
                utilities_budget = self.monthly_salary * 0.15
                groceries_budget = self.monthly_salary * 0.15
                transportation_budget = self.monthly_salary * 0.15
                entertainment_budget = self.monthly_salary * 0.05
                savings_budget = self.monthly_salary * 0.1
                debt_payment_budget = self.monthly_salary * 0.15  

                print("\nSuggested Budget Allocations (With Debt) based off of monthly salary:")  # prints the suggested budget allocations with debt
                print(f"Rent or Mortgage: ${rent_mortgage_budget:.2f}")
                print(f"Utilities (Electricity, Water, Gas): ${utilities_budget:.2f}")
                print(f"Groceries: ${groceries_budget:.2f}")
                print(f"Transportation (Car payments, Fuel, Public Transit): ${transportation_budget:.2f}")
                print(f"Entertainment: ${entertainment_budget:.2f}")
                print(f"Savings: ${savings_budget:.2f}")
                print(f"Debt Payment: ${debt_payment_budget:.2f}")
                break  # quits the budget creator after it prints the suggestions
            elif bchoice==2:
                break
            else:
                print("Invalid number, please choose a number between (1-2)")
                
           
     
```

```python
user= budget_manager()

while True:
    print("\nWelcome to the Budget Manager")
    print("1: Set Salary")
    print("2: Set Expenses")
    print("3: Show Expenses")
    print("4: Create a Sample Budget based off of your salary")
    print("5: Quit") 
    choice=int(input("Select a Number (1-5): "))
    
    if choice==1:
        user.set_salary()
    elif choice==2:
        user.set_expenses()
    elif choice==3:
        user.show_expenses()
    elif choice==4:
        user.sample_budget()
    elif choice==5:
        break    
    else:
        print("Invalid number, please choosee a number between (1-5)")
```
